```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image 
```

```python
image = cv2.imread("rocket.webp", cv2.IMREAD_COLOR)
plt.imshow(image[:,:,::-1])
```

```python
image_with_line = image.copy()

cv2.line(image_with_line, (200, 100), (400, 100), (0,255,255), thickness=5, lineType=cv2.LINE_AA)
plt.imshow(image_with_line)
```
