```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image 
```

```python
img_bgr = cv2.imread("nz.jpg", cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
Image(filename="nz.jpg")
```

```python
matrix = np.ones(img_rgb.shape, dtype = "uint8") * 50

img_rgb_bright = cv2.add(img_rgb, matrix)
img_rgb_dark= cv2.subtract(img_rgb, matrix)

plt.figure(figsize=[18,5])
plt.subplot(131);plt.imshow(img_rgb_bright);plt.title(("Brighter"))
plt.subplot(132);plt.imshow(img_rgb_dark);plt.title(("Dark"))


# for changing contrast, multiply by 0.8 and 1.2 and multiply matrix and image with keeping type casting in mind
# however it might happen that some pixels with value 255 go above 255 when multiplied by 1.2, in that case use cv2.clip()
```

```python
# image thresholding

piano_image = cv2.imread("piano.jpg", cv2.IMREAD_GRAYSCALE)
retval, img_thresh_gbl_1 = cv2.threshold(piano_image, 50, 255, cv2.THRESH_BINARY)
retval, img_thresh_gbl_2 = cv2.threshold(piano_image, 130, 255, cv2.THRESH_BINARY)
img_thresh_adp = cv2.adaptiveThreshold(piano_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 7)

plt.figure(figsize=[18,15])
plt.subplot(221); plt.imshow(piano_image, cmap="gray"); plt.title("Original");
plt.subplot(222); plt.imshow(img_thresh_gbl_1, cmap="gray"); plt.title("Thresholded (global: 50)");
plt.subplot(223); plt.imshow(img_thresh_gbl_2, cmap="gray"); plt.title("Thresholded (global: 100)");
plt.subplot(223); plt.imshow(img_thresh_gbl_2, cmap="gray"); plt.title("Thresholded (global: 100)");
plt.subplot(224); plt.imshow(img_thresh_adp, cmap="gray"); plt.title("Thresholded (adaptive)");

# simplifies an image on a grayscale or color image into a binary image, by dividing pixels into two classes on a threshold value
# Thresholding converts an image into a binary format by assigning a value (usually 0 for black or background and 255 for white or foreground) to each pixel based on its intensity. 
```
