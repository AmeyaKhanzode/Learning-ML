```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image       #display and render images in the notebook
```

```python
# display an image
Image(filename="checkerboard.png")
```

```python
# read image as grey scale
cb_image = cv2.imread("checkerboard.png", 0)    # 0 is a flag that displays grey scale. 1 for colour. -1 for alpha channel.
np.set_printoptions(threshold=np.inf)

print(cb_image)
# as we can see below, its a numpy array that represents the image by showing the intensity of the colour on the grey scale.
```

```python
print("image size: ", cb_image.shape)
print("image dtype: ", cb_image.dtype)
```

```python
plt.imshow(cb_image)    # can add cmap=gray to get a black and white representation
```
