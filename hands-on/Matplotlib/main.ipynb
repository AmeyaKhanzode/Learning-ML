```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# scatter plot - good for seeing population density or for coordinates on a map.

x_data = np.random.random(2000) * 100
y_data = np.random.random(2000) * 100

plt.figure(figsize=(5, 3))
plt.scatter(x_data, y_data, alpha=0.3)
'''
just plots random points (x,y), alpha parameter shows dark spots where a lot of points are overlapping.
this helps in looking for clusters.
''' 
```

```python
# line graph - good for showing the variation of one thing over some time.

years = [2005 + x for x in range(10)]
weights = [67, 64, 62, 69, 71, 63, 63, 63, 66, 64]

plt.figure(figsize=(5, 3))
plt.plot(years, weights)            #linestyle="--" for dotted line
```

```python
# bar graph - good for comparing multiple things with one parameter.

x = ["C++", "Python", "C", "java", "Javascript"]
y = [10, 9, 24, 17, 4]

plt.figure(figsize=(5, 3))
plt.bar(x, y)
```

```python
ages = np.random.normal(20, 1.5, 1000)      # 1000 values with mean 20 and sd 1.5
plt.figure(figsize=(5, 3))
plt.hist(ages, bins=300);
```
